# Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Import dan Baca Gambar

```python
#Rama Dinantiar (202331044)
img = cv2.imread("Rama Backlight.jpg")
```

### Visualisasi Gambar Asli

```python
image_backlight = cv2.imread('Rama Backlight.jpg')
plt.imshow(cv2.cvtColor(image_backlight, cv2.COLOR_BGR2RGB))
plt.title('Gambar Asli')
plt.axis('off')
plt.show()
```

### Transformasi Warna RGB ke Skala Abu-Abu

```python
gray = cv2.cvtColor(image_backlight, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Gambar Grayscale')
plt.axis('off')
plt.show()
```

### Mencerahkan Gambar Grayscale Secara Manual dengan Iterasi Piksel

```python
# Ambil ukuran gambar
baris, kolom = gray.shape

# Cerahkan secara manual
beta = 70
citra_cerah = np.zeros((baris, kolom), dtype=np.uint8)

for x in range(baris):
    for y in range(kolom):
        nilai = gray[x, y] + beta
        citra_cerah[x, y] = np.clip(nilai, 0, 255)

# Tampilkan hasil
plt.imshow(citra_cerah, cmap='gray')
plt.title('Citra Grayscale Dicerahkan')
plt.axis('off')
plt.show()
```

### Mengubah Kontras Gambar Grayscale Secara Manual dengan Operasi Per-Piksel

```python
# Ambil ukuran gambar
baris, kolom = gray.shape

# Cerahkan secara manual
alpha = 1
citra_kontras = np.zeros((baris, kolom), dtype=np.uint8)

for x in range(baris):
    for y in range(kolom):
        nilai = gray[x, y] * alpha
        citra_kontras[x, y] = np.clip(nilai, 0, 255)

# Tampilkan hasil
plt.imshow(citra_kontras, cmap='gray')
plt.title('Citra Grayscale DiKontraskan')
plt.axis('off')
plt.show()
```

### Mengubah Kecerahan Kontras dan Gambar Grayscale

```python
# Ubah ke float untuk akurasi saat perhitungan kontras
citra_cerah_float = citra_cerah.astype(np.float32)

# Tentukan faktor kontras (alpha)
alpha = 1  # >1 = lebih kontras, <1 = kurang kontras

# Terapkan peningkatan kontras manual
citra_kontras = (citra_cerah_float - 128) * alpha + 128
citra_kontras = np.clip(citra_kontras, 0, 255).astype(np.uint8)

# Tampilkan hasil
plt.imshow(citra_kontras, cmap='gray')
plt.title('Citra Grayscale Dicerahkan dan Dikontraskan')
plt.axis('off')
plt.show()
```
